```python
import pandas as pd
import json
```

```python
train = json.load(open('train.json',))
len(train)
```

```python
# Cuisines KÃ•IK
cuisines = []
for i in range(len(train)):
  cuisines.append(train[i]["cuisine"])
```

```python
# Cuisines Unique ja arvud
unique = list(set(cuisines))
#len(myset)

distr = []
for c in unique:
  coun = len([1 for i in cuisines if i==c])
  distr.append(coun)
```

```python
# Sorteerime
sorted_indexes = [index for index, value in sorted(enumerate(distr), reverse=True, key=lambda x: x[1])]
distr[:] = [distr[i] for i in sorted_indexes] 
#u = list(X_train.columns)
unique[:] = [unique[i] for i in sorted_indexes] 
```

```python
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt

#objects = ('Python', 'C++', 'Java', 'Perl', 'Scala', 'Lisp')
y_pos = np.arange(len(unique))
#performance = [10,8,6,4,2,1]

plt.barh(y_pos, distr, align='center', color="#2C5696")
plt.yticks(y_pos, unique)
plt.title('Distribution of cuisines')
#plt.figure(figsize = (10,10))

# save the figure
figure = plt.gcf()

figure.set_size_inches(4, 6)
plt.savefig('plot_color_TU_darkblue.png', bbox_inches='tight')#, dpi=100)#), width=10, height=100)# dpi=100

plt.show()
```
